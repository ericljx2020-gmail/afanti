```python
import json
from multiagent_func import run
```

```python
with open("problem_set.json", 'r') as file:
    data = json.load(file)
```

```python
record_all = []
for ps in data[:6]:
    p1 = ps['similar_question_1']
    p2 = ps['original_question']
    p3 = ps['original_explanation']
    cur = run(p1,p2,p3, "五年级")
    cur = [p1] + cur
    record_all.append(cur) 
```

```python
record_all
```

```python
with open("sample.md", 'w') as file:
    for record in record_all:
        file.write("## 题目：\n")
        file.write(repr(record[0]) + '\n\n')
        file.write("## 板书：\n")
        file.write(repr(record[1]) + '\n\n')
        file.write("## 讲解：\n")
        file.write(repr(record[2]) + '\n\n')
```
